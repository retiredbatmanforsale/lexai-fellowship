```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
# Function to plot cumulative explained variance
def PCA_EigVector_Plot(scalarData):
    pca = PCA()
    pca.fit(np.array(scalarData).T.astype(object))
    cumsum = np.cumsum(pca.explained_variance_ratio_)*100
    d = [n for n in range(len(cumsum))]
    plt.figure(figsize=(10, 10))
    plt.plot(d, cumsum, color='red', label='cumulative explained variance')
    plt.title('Cumulative Explained Variance as a Function of the Number of Components')
    plt.ylabel('Cumulative Explained variance')
    plt.xlabel('Principal components')
    plt.axhline(y=95, color='k', linestyle='--', label='95% Eigen energy')
    plt.axhline(y=99, color='k', linestyle='--', label='99% Eigen energy')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()
    
    # Print the number of components needed for different thresholds
    comp_95 = next(i for i, val in enumerate(cumsum) if val >= 95) + 1
    comp_99 = next(i for i, val in enumerate(cumsum) if val >= 99) + 1
    print(f"Number of components needed for 95% variance: {comp_95}")
    print(f"Number of components needed for 99% variance: {comp_99}")
    
    # Return explained variance ratio for reference
    return pca.explained_variance_ratio_
```

```python
# Example 1: Iris Dataset
print("Example 1: Iris Dataset")
# Load iris dataset
iris = load_iris()
X = iris.data
feature_names = iris.feature_names
print(X)
print(feature_names)
```

```python
# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# Call our function with the scaled data (note: we use X_scaled.T to match the function's expectation)
print("Explained variance ratio for each component:")
explained_variance = PCA_EigVector_Plot(X_scaled.T)
print(explained_variance)


for i, var in enumerate(explained_variance):
    print(f"PC{i+1}: {var*100:.2f}%")
```

```python
# Example 2: Generated Dataset with more features
print("\nExample 2: Generated Dataset with more dimensions")
# Generate synthetic data with strong correlation between some features
np.random.seed(42)
n_samples = 200
n_features = 20
```

```python
# Create a dataset where the first few dimensions have most of the variance
X_synth = np.random.randn(n_samples, n_features)
# Add some structure: first 3 features are highly correlated
X_synth[:, 0] = 3 * X_synth[:, 1] + 2 * X_synth[:, 2] + np.random.randn(n_samples) * 0.5
X_synth[:, 3] = -2 * X_synth[:, 4] + X_synth[:, 5] + np.random.randn(n_samples) * 0.3
```

```python
# Scale features differently to create variance differences
for i in range(n_features):
    X_synth[:, i] *= np.exp(-i/5)  # Exponential decay of feature variance
```

```python
# Scale the data
X_synth_scaled = scaler.fit_transform(X_synth)
```

```python
# Call our function
print("Explained variance ratio for each component:")
explained_variance = PCA_EigVector_Plot(X_synth_scaled.T)
# Print first 10 components for brevity
for i, var in enumerate(explained_variance[:10]):
    print(f"PC{i+1}: {var*100:.2f}%")
```

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Step 1: Load MNIST dataset
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist.data, mnist.target

# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA to preserve 95% of the variance
pca = PCA(n_components=0.95)  # Keep 95% of total variance
X_pca = pca.fit_transform(X_scaled)

# Step 4: Output info
print(f"Original shape: {X.shape}")
print(f"Reduced shape after PCA: {X_pca.shape}")
print(f"Number of components chosen: {pca.n_components_}")

# Optional: Plot explained variance ratio
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), 
         pca.explained_variance_ratio_.cumsum(), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA on MNIST (95% Variance)')
plt.grid(True)
plt.show()
```
