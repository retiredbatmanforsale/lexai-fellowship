üé≤ PROBLEM SETUP: Sum of Two Dice

You‚Äôre rolling two fair 6-sided dice (Die A and Die B). Each die has numbers from 1 to 6.

You want to calculate the probability distribution of their sum:
	‚Ä¢	Minimum sum = 1 + 1 = 2
	‚Ä¢	Maximum sum = 6 + 6 = 12

You want to know the probability of each sum from 2 to 12.

‚∏ª

üî¢ Step 1: Represent Each Die as a Probability Vector

Since each die is fair:

Each die has a discrete uniform distribution over values 1 to 6.

So we can represent the probability of each outcome as a vector:

Die A (and Die B) probabilities:
Index:      0     1     2     3     4     5
Face:       1     2     3     4     5     6
P(value):  1/6   1/6   1/6   1/6   1/6   1/6

Let‚Äôs define this as a vector a:

a = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

Same for the second die b.

‚∏ª

üîÅ Step 2: What is Convolution?

The convolution operation combines these two vectors to produce a new vector c such that:

c[n] = ‚àë a[k] * b[n - k]

This is summing over all the ways to combine values from Die A and Die B to get a total sum.

But since this is discrete and finite, we do discrete convolution of two sequences a and b:

c = a * b  ‚Üê (discrete convolution)

This c gives us probabilities for each possible sum, shifted appropriately.

‚∏ª

üß† Intuition: What Does Convolution Do?

When you convolve two probability distributions (like the two dice), the result is a distribution of their sum.

That is exactly what we want.

‚∏ª

üî¢ Step 3: Perform Discrete Convolution

We convolve:

a = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
b = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

The result c will have length 6 + 6 - 1 = 11 elements:

Each index i in the result corresponds to the sum = i + 2.

Now compute it:

Let‚Äôs write it as sum probabilities:

Sum	Ways	Probability
2	(1,1)	1/36
3	(1,2), (2,1)	2/36
4	(1,3), (2,2), (3,1)	3/36
5	(1,4), (2,3), (3,2), (4,1)	4/36

8	‚Ä¶	5/36
9	‚Ä¶	4/36
10	‚Ä¶	3/36
11	‚Ä¶	2/36

That‚Äôs the result of convolution.

So:

c = [1/36, 2/36, 3/36, 4/36, 5/36, 6/36, 5/36, 4/36, 3/36, 2/36, 1/36]

üîÅ Step 4: Match It to CNN Convolution

In CNNs, you are convolving a filter (kernel) with the input (like an image patch).
	‚Ä¢	Dice example:
	‚Ä¢	a = input distribution (Die A)
	‚Ä¢	b = filter (Die B)
	‚Ä¢	a * b = convolved output (sum distribution)

This is exactly what CNN does, except:
	‚Ä¢	In images: the input is a 2D grid of pixel values
	‚Ä¢	The kernel is usually a small 2D filter (e.g., 3x3)
	‚Ä¢	The convolution aggregates local neighborhoods (just like dice values aggregate sum probabilities)

So convolution in CNNs generalizes this same ‚Äúoverlapping and summing‚Äù idea across image regions.

ü§ñ Why is this Important for CNNs?

This example reinforces a key idea:

Convolution is a way to combine two sets of values ‚Äî weighting one against the other across all alignments.

In CNNs:
	‚Ä¢	Input = image patches (e.g., intensity values, like a)
	‚Ä¢	Kernel = filter weights (like b)
	‚Ä¢	Output = feature map (like c), which highlights how strongly that patch matches the filter

And just like our dice example, each value in the output is a weighted combination of input values, where the weights are defined by the kernel.
